```python
def f(x):
    return (1+ math.cos(x))**2 * math.sin(2*x) - 3.015544498
```

```python
def bisection(a,b):
    tol = 23 
    c = (b-a)/2
    i =0

    while i<=tol:
        print(i)
        print("a-->{:.6f} |f(a)-->{:.6f}".format(a, f(a)))
        print("b-->{:.6f} |f(b)-->{:.6f}".format(b, f(b)))
        print("c-->{:.6f} |f(c)-->{:.6f}".format(c, f(c)))
        if f(c) == 0:
            break
        elif f(a)*f(c)<0:
            b = c
        else:
            a = c
        c = (b+a)/2
    
        i+=1
    print(i)
    print("a-->{:.6f} |f(a)-->{:.6f}".format(a, f(a)))
    print("b-->{:.6f} |f(b)-->{:.6f}".format(b, f(b)))
    print("c-->{:.6f} |f(c)-->{:.6f}".format(c, f(c)))
    print(math.degrees(c))
    return c
    
```

```python
print(bisection(math.radians(0),math.radians(40)))
```
